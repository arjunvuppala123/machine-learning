```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
```

```python
house_price=[245,312,279,308,199,219,405,324,310,255]
size=[1400,1600,1700,1825,1100,1550,2550,2450,1450,1700]
```

```python
size1=np.array(size).reshape((-1,1))
```

```python
regr=linear_model.LinearRegression()
regr.fit(size1,house_price)
print(regr.coef_)
print(regr.intercept_)
```

```python
def graph(formula,x_range):
    x=np.array(x_range)
    y=eval(formula)
    plt.plot(x,y)
    
```

```python
graph('regr.coef_*x + regr.intercept_',range(1000,4000))
plt.scatter(size,house_price,color='black')
plt.ylabel('House Price')
plt.xlabel('size of house')
plt.show()
```
